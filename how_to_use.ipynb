### 1. Getting Telegram API creadentials

1. Go to Telegram API website **https://my.telegram.org/apps**
2. **Login** with your Telegram phone number
3. **Create new application:**
   - App title: `My Telegram Parser` (any name)
   - Short name: `parser` (any short name)  
   - Platform: `Desktop`
4. **Copy your credentials:**
   - API ID (numbers): `1234567`
   - API Hash (string): `abc123def456...`
5. Create a file named `.env` in your project folder with your credentials:
   
   TG_API_ID=`1234567` (should be integer, not a string!)  
   TG_API_HASH=`abc123def456...`

### 3. Finding the chats and channels ids

#### Method 1: Using Telegram Bots

1. **For any chat**: Forward a message from the target chat to **@userinfobot**
  - It will reply with chat information including the ID

2. **For channels you admin**: Add **@raw_data_bot** to your channel
  - Send any message, and it will show the channel ID

#### Method 2: Using Telegram web interface


1. **Login into web interface**
  - Login to **https://web.telegram.org**
  - Click on the chat or channel you want to parse
  - Look at the URL in your browser address bar

2. **Extract the ID from URL:**
  - For **public channels/groups**: URL shows `https://web.telegram.org/k/#@channel_username`
    - Use: `"@channel_username"` (include the @ symbol)
  - For **personal chats**: URL shows `https://web.telegram.org/k/#1234567890`
    - Use: `1234567890` (positive number)
  - **private chats/groups** are currently not avaliable for downloads 

### 4. Parcing API data

```python
# Replace with your own chat_id
chat_id = '@besedernedaber'
```

```python
# Initializing the parser.
parser = TgApiParser(chat_id=chat_id)
```

```python
# Loading data from the specified chat with a limit on the number of messages.

# The first time you launch this line you will be asked
# to enter your phone number used for Telegram registation,
# and then to enter the code sent to you via Telegram.
# After that, the data will be loaded and saved to a file.

await parser.load_data()
```

```python
# Parsing the data received by API
parser.extract_messages()
```

```python
# Load media files (photos, videos, documents, voice messages, audio files to a folder)

await parser.get_files()
```

```python
# Save the parsed messages to a CSV file(s) for further analysis
parser.save_chat(save_actions=True)
```

### 5. Use cases

```python
import pandas as pd
```

```python
# Extract the parsed content messages from the parser
# (messages from the users, not system messages)
content_messages = parser.content_messages

# Convert the list of message dictionaries to a pandas DataFrame for analysis
content_messages_table = pd.DataFrame(content_messages)

# Display the first 10 rows to inspect the data structure and content
content_messages_table.head(10)
```

```python
# What are the most reposted channels?

content_messages_table.loc[content_messages_table['forward_from'] == 'chat', 'forward_name'].value_counts()
```

```python
#  Get non- text media messages with links to stored files 
content_messages_table.loc[content_messages_table.file.notna(), ['msg_id', 'posting_time', 'media_type', 'content', 'file_name', 'file']]
```
